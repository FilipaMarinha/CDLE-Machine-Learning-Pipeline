# **Introduction** <a name = "intro"></a>
[[go back to top]](#index)

In the current paradigm of digital health, Intensive Care Units (ICUs) daily generate massive volumes of complex and heterogeneous data. The processing and analysis of this information at scale, commonly referred to as Big Data, offers an unprecedented opportunity to optimize hospital management and improve clinical prognosis through predictive Machine Learning models.

This project focuses on the predictive analysis of the Length of Stay (LOS) for ICU patients, utilizing the MIMIC-III (Medical Information Mart for Intensive Care) dataset. MIMIC-III is a publicly accessible database containing de-identified data from over 40,000 patients, including vital signs, medications, laboratory measurements, and clinical notes. The dataset includes files exceeding 4 GB in compressed format, with critical tables such as CHARTEVENTS surpassing 300 million records.

The primary challenge of this work transcends traditional statistical modeling, lying in the necessity of implementing efficient processing pipelines capable of handling high data volume without compromising computational resources. To this end, we explore and compare different distributed and parallel processing tools, such as PySpark, Dask, and Polars, evaluating the trade-off between execution time (latency) and predictive accuracy.

The central objective is to address the following question: Is it possible to accurately predict a patient's length of stay based on their initial clinical events using a scalable Big Data architecture? To achieve this, the work is structured as follows:

- **Ingestion and Cleaning**: Strategies for the efficient loading of massive datasets.

- **Feature Engineering**: Aggregation of vital signs per patient and definition of temporal observation windows.

- **Modeling and Evaluation**: Implementation of regression/classification algorithms and a critical analysis of the performance of the tools utilized.

# **The Project** <a name = "project"></a>
[[go back to top]](#index)

The main goal of this project is to build a machine learning pipeline capable of handling large-scale healthcare datasets. For this purpose, we use the CHARTEVENTS table from the MIMIC-III Clinical Database, which contains time-stamped records of clinical events collected from patients during their stay in the Intensive Care Unit (ICU). The dataset is highly large-scale and complex, with the compressed file occupying approximately 4.2 GB, and a significantly larger size when uncompressed (34 GB). Its volume, temporal structure, and heterogeneity introduce challenges related to scalability and efficient data processing in machine learning pipelines.

The objective of this project is to design and implement a complete machine learning workflow, including data preprocessing, feature engineering, model training and validation, performance evaluation, and result interpretation. Appropriate tools such as (TOOLS) will be used to efficiently process and analyze the data.

The project is divided into two main tasks:

- **Exploratory Data Analysis:** statistical analysis and visualization of ICU patient data, grouped by SUBJECT_ID and HADM_ID. This includes data sampling, cleaning, temporal analysis of clinical events, and visualization of key variables over time.
- **Machine Learning-Based Prediction:** prediction of ICU Length of Stay (LOS) using data from the first 24 hours of patient admission. This task includes feature extraction, aggregation of time-series data, model training, hyperparameter tuning, and evaluation using appropriate regression metrics.

The final objective is to predict ICU Length of Stay based on early clinical measurements, supporting decision-making in critical care environments while ensuring proper handling of temporal data.
